### 1) 0 / 1 Knapsack problem ( recursive )

```python
def knapsack_recursive(wt, val, W, n):
    if n==0 or W==0:
        return 0
    if wt[n-1]<=W:
        inc = val[n-1]+knapsack_recursive(wt,val,W-wt[n-1],n-1)
        exc = knapsack_recursive(wt,val,W,n-1)
    else:
        return knapsack_recursive(wt,val,W,n-1)
    return max(inc,exc)

weights = [2, 3, 4, 5]
values = [3, 4, 5, 6]
capacity = 5
n = len(values)

max_value = knapsack_recursive(weights, values, capacity, n)
print(f"Maximum value in knapsack of capacity {capacity}: {max_value}")
```

### 0 / 1 Knapsack problem ( Memoisation )

```python
def knapsack_memoisation(wt, val, W, n, dp):
    if n==0 or W==0:
        return 0
    if dp[n][W]!=-1:
        return dp[n][W]
    if wt[n-1]<=W:
        include = val[n-1]+knapsack_memoisation(wt,val,W-wt[n-1], n-1, dp)
        exclude = knapsack_memoisation(wt,val,W, n-1, dp)
        dp[n][W] = max(include, exclude)
    else:
        return knapsack_memoisation(wt,val,W, n-1, dp)

    return dp[n][W]


weights = [2, 3, 4, 5]
values = [3, 4, 5, 6]
capacity = 5
n = len(values)

dp = [[-1 for _ in range(capacity+1)] for _ in range(n+1)]

max_value = knapsack_memoisation(weights, values, capacity, n, dp)
print(f"Maximum value in knapsack of capacity {capacity}: {max_value}")
```

```python
import time
import os

def knapsack_animated(weights, values, capacity, delay=0.5):
    n = len(values)
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]

    def clear():
        os.system('cls' if os.name == 'nt' else 'clear')

    def print_dp_table(dp, highlight=None):
        print("     " + "  ".join([f"{i:>2}" for i in range(len(dp[0]))]))
        print("    " + "---" * len(dp[0]))
        for i, row in enumerate(dp):
            row_str = ""
            for j, cell in enumerate(row):
                if highlight == (i, j):
                    row_str += f"\033[92m{cell:>3}\033[0m"  # Green highlight
                else:
                    row_str += f"{cell:>3}"
            print(f"{i:>2} | {row_str}")

    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            clear()
            print(f"🔄 Considering item {i} (weight={weights[i - 1]}, value={values[i - 1]})")
            if weights[i - 1] <= w:
                include = values[i - 1] + dp[i - 1][w - weights[i - 1]]
                exclude = dp[i - 1][w]
                dp[i][w] = max(include, exclude)
                decision = "✅ Included" if include > exclude else "❌ Excluded"
                print(f"Capacity {w}: max({exclude}, {values[i - 1]} + {dp[i - 1][w - weights[i - 1]]}) → {dp[i][w]} → {decision}")
            else:
                dp[i][w] = dp[i - 1][w]
                print(f"Capacity {w}: Item too heavy, carry forward {dp[i][w]}")

            print("\n🧩 DP Table:")
            print_dp_table(dp, highlight=(i, w))
            time.sleep(delay)

    clear()
    print("✅ Final DP Table:")
    print_dp_table(dp)
    print(f"\n🎯 Maximum value in knapsack of capacity {capacity}: {dp[n][capacity]}")
    return dp[n][capacity]
```

```python
weights = [2, 3, 4, 5]
values = [3, 4, 5, 6]
capacity = 5

knapsack_animated(weights, values, capacity, delay=0.7)
```
